```python
import pandas as pd
import numpy as np
import zipfile
```

```python
#lendo os arquivos compactados direto da url usando o zipfile

dados2018 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download', encoding='latin1', compression ='zip', sep=';')
dados2019 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download', encoding='latin1', compression ='zip', sep=';')
dados2020 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download', encoding='latin1', compression='zip', sep=';')

#concatenando os 3 dfs em um só

dadosconc = pd.concat([dados2018,dados2019,dados2020])
dadosconc
```

```python
# questão 2a. Usando o len do df posso conseguir a quantidade de linhas que correspondem a quantidade de acidentes

print('Em 2018 ocorreram', len(dados2018), 'acidentes, em 2019 foram', len(dados2019), 'acidentes e', len(dados2020), 'em 2020.')
```

```python
# questão 2b. Também usando o len posso conseguir a quantidade de variaveis, mas dessa  vez usando o .columns para contar as colunas do df

print('O df de 2018 possui', len(dados2018.columns), 'variávies,', 'o de 2019 possui', len(dados2019.columns), 'e o df de 2020 possui', len(dados2020.columns), 'variáveis. Ou seja, possuem a mesma quantidade.')
```

```python
# questão 2c. Usando o groupby é possível agrupar os dados pela coluna que quero, usando também o reset_index para resetar e adicionar um contador ao lado.

print('Em 2020, as cidades que mais ocorreram acidentes foram:')
dados20 = dados2020.groupby(['municipio']).size().reset_index(name = 'count')
dados20 = dados20.sort_values(by = 'count', ascending=False)
dados20.head()
```

```python
# questão 2d. Utilizo a filtragem para obter os referentes a PB 

pb2019 = dados2019[(dados2019.uf == 'PB') & (dados2019.feridos_graves >= 1)]
print('A Paraiba obteve', len(pb2019), 'acidentes com pessoas gravemente feridas em 2019')
pb2019.head()
```

```python
# questão 3a. Como jpa concatenei tudo na questão 1, apenas filtrarei com o groupby e depois organizar os rankings para salvar o csv

print("Estados com mais acidentes entre 2018 e 2020:")
lista = dadosconc.groupby(['uf']).size().reset_index(name='total')
lista1 = lista.sort_values(by='total', ascending=False )
lista1.to_csv("totalporeEstado.csv")
lista1
```

```python
# questão 3b. Mesmo processo da anterior mas salvando uma cópia no final

print('Quantidade de acidentes por dia da semana')
lista2 = dadosconc.groupby(['dia_semana']).size().reset_index(name='total')
lista2cop = lista2.sort_values(by = 'total', ascending = False)
print(lista2cop)
```

```python
ranking_dia = lista2cop.copy()
ranking_dia.to_csv("rankingDia.csv")
```

```python
#questão 4a. O procedimento é parecido com o da questão 3 mas vou utillizar o tail para me mostrar a última linha  que representa a causa mais rara

ranking = dadosconc.groupby(['causa_acidente']).size().reset_index(name = 'total')
rankingcop = ranking.sort_values(by='total', ascending = False)
print("A causa mais frequente foi: ")
rankingcop.head(1)
```

```python
print("E a causa menos frequente foi:")
rankingcop.tail(1)
```

```python
#questão 4b.

#faço a soma dos valores correspondentes as colunas que vou usar

soma_pes20 =  dados2020['pessoas'].sum()
soma_ile20 = dados2020['ilesos'].sum()
soma_grave20 = dados2020['feridos_graves'].sum()

#faço a media dividindo os feridos gravemente e ilesos pelo total de pessoas

media_ile = round(soma_ile20/soma_pes20,2)
media_graves = round(soma_grave20/soma_pes20,2)

#printo os dados adquiridos

print('Em 2020 a proporção entre feridos e ilesos foi aproximadamente', media_ile)
print('já a proporção entre gravemente feridos e ilesos foi cerca de', media_graves,'\n')

#repito os passos com os outros anos:

#2019:
soma_pes19 = dados2019['pessoas'].sum()
soma_ile19 = dados2019['ilesos'].sum()
soma_grave19 = dados2019['feridos_graves'].sum()
media_ile19 = round(soma_ile19/soma_pes19,2)
media_graves19 = round(soma_grave19/soma_pes19,2)
print('Em 2019 a proporção entre feridos e ilesos foi aproximadamente', media_ile19)
print('Já a proporção entre gravemente feridos e ilesos foi cerca de', media_graves19, '\n')

#2018:
soma_pes18 = dados2018['pessoas'].sum()
soma_ile18 = dados2018['ilesos'].sum()
soma_grave18 = dados2018['feridos_graves'].sum()
media_ile18 = round(soma_ile18/soma_pes18,2)
media_graves18 = round(soma_grave18/soma_pes18,2)
print('Em 2018 a proporção entre feridos e ilesos foi aproximadamente', media_ile18)
print('já a proporção entre gravemente feridos e ilesos foi cerca de', media_graves18, '\n')
```

```python
#questão 4c. criei copias dos dataframes que vinha usando e os concatenei em um novo, usei o grupby, reset index e printei do maior para o menor

cop_dados18 = dados2018.copy()
dados18 = cop_dados18.assign(ano=2018)

cop_dados19 = dados2019.copy()
dados19 = cop_dados19.assign(ano=2019)

cop_dados20 = dados2020.copy()
dados20 = cop_dados20.assign(ano=2020)

concatenados = pd.concat([dados18, dados19, dados20])
datafr = concatenados.groupby(['ano']).size().reset_index(name='total')
data_concat = datafr.sort_values(by='total', ascending=False)

print('Esses são os totais aproximados de acidentes que ocorreram a cada ano, onde é possível ver uma leve queda no ano de 2020, logo, a pandemia ajudou a diminuir a quantidade de acidentes')
data_concat
```
